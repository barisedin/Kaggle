# Read the data

```python
train= pd.read_csv("/kaggle/input/playground-series-s6e1/train.csv") 

test_org = pd.read_csv("/kaggle/input/playground-series-s6e1/test.csv") 
test=test_org.copy()

def drop_val(data):
        data.drop(columns=["id"],inplace=True)


drop_val(train)
drop_val(test)
```

# Encoding data

```python
#araliklar = [16, 19, 21, 22,24]
#etiketler = [0, 1, 2, 3]
#
#df['age'] = pd.cut(df['age'], bins=araliklar, labels=etiketler).astype(int)
#
#print("Belirli aralıklarla gruplama:\n")
#print(df)



def encode_age (data):
    labels=[0, 1, 2, 3]
    intervals=[16, 19, 21, 22,24]
    
    data['age'] = pd.cut(data['age'], bins=intervals, labels=labels).astype(int)

encode_age(train)
encode_age(test)
```

```python
encode_columns=['course_b.com','course_b.sc',
       'course_b.tech', 'course_ba', 'course_bba', 'course_bca',
       'course_diploma', 'study_method_coaching', 'study_method_group study',
       'study_method_mixed', 'study_method_online videos',
       'study_method_self-study','goodstu']


def data_create(data):
    data["total_effort"] = (data["study_hours"] * data["class_attendance"])/10
    data["sleepquality"] = ((data["sleep_hours"]>=7) & (data["sleep_quality"]==2))
    data["goodstu"] = ((data["class_attendance"] > 80) & (data["study_hours"] > 5)).astype(int)

    return data


def one_encode (data):
    data = pd.get_dummies(data, columns=["course","study_method"])
    
    data["internet_access"] = data["internet_access"].apply({"yes": 1, "no": 0}.get)
    data["gender"] = data["gender"].apply({"female": 1, "male": 0, "other":2}.get)
    data["sleep_quality"]=data["sleep_quality"].apply({"poor":0,"average":1,"good":2}.get)
    data["facility_rating"]=data["facility_rating"].map({"low":0,"medium":1,"high":2}.get)
    data["exam_difficulty"]=data["exam_difficulty"].map({"easy":0,"moderate":1,"hard":2}.get)

    

    data[encode_columns] = data[encode_columns].map({True: 1, False: 0}.get)

    return data


data_create(train)
data_create(test)

train = one_encode(train)
test  = one_encode(test)



```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from catboost import CatBoostRegressor # Kütüphaneyi eklemeyi unutma

# K-Fold Ayarları
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# X ve y Hazırlığı
X = df.drop(columns="exam_score", axis=1)
y = df["exam_score"]

# Sonuçları saklayacağımız boş diziler
cat_oof_preds = np.zeros(len(X))       
cat_test_preds = np.zeros(len(test))   
cat_scores = []


# İndeksleri sıfırla
X = X.reset_index(drop=True)
y = y.reset_index(drop=True)

# Eğer kategorik sütunların varsa buraya indekslerini yazmalısın.
# Eğer verilerin hepsi sayısal ise bu listeyi boş bırakabilirsin.
# Örnek: cat_features_indices = [0, 2, 5] 
cat_features_indices = [] 

print("K-Fold Eğitimi Başlıyor...\n")


cb_params = {
    'bagging_temperature': 0.006952130531190703,
    'border_count': 254,
    'depth': 6,
    'iterations': 1148,
    'l2_leaf_reg': 5,
    'learning_rate': 0.1484872065780541,
    'random_strength': 3.6941233379852148
}

#cb_params = {
#    'bagging_temperature': 6.4791075480233555,
#    'border_count': 254,
#    'depth': 7,
#    'iterations': 3582,
#    'l2_leaf_reg': 3.253942650188866,
#    'learning_rate': 0.029894716076568403,
#    'random_strength': 8.5991320623756
#}



for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
    
    # --- A. VERİYİ BÖLME ---
    X_tr, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_tr, y_val = y.iloc[train_idx], y.iloc[val_idx]
    
    # --- B. SCALING (Sadece Linear Model İçin) ---
    scaler = StandardScaler()
    X_tr_scaled = scaler.fit_transform(X_tr) # Sadece sayısal sütunlar olmalı!
    X_val_scaled = scaler.transform(X_val)
    
    # Test setini (kaggle submission) dönüştürürken dikkat:
    # Linear model 'kagtest_scaled' kullanacak, CatBoost orijinal 'test' dataframe'ini kullanacak.
    kagtest_scaled = scaler.transform(test) 
    
    # --- C. MODEL 1: LINEAR REGRESSION ---
    lin_model = LinearRegression()
    lin_model.fit(X_tr_scaled, y_tr)
    
    # Linear Tahminler
    p_lin_train = lin_model.predict(X_tr_scaled) 
    p_lin_val = lin_model.predict(X_val_scaled)   
    p_lin_test = lin_model.predict(kagtest_scaled)
    
    # --- D. RESIDUAL HESAPLAMA (KRİTİK BÖLÜM) ---
    # CatBoost'un öğreneceği şey: Gerçek Değer - Linear Tahmin
    residuals_tr = y_tr - p_lin_train
    
    # BURASI EKSİKTİ: Validation için de residual hesaplamalısın!
    # Çünkü CatBoost 'residuals' tahmin etmeye çalışıyor, 'exam_score' değil.
    residuals_val = y_val - p_lin_val 

    # --- E. MODEL 2: CATBOOST (Residuals üzerine) ---
    model = CatBoostRegressor(
    **cb_params,
    loss_function='RMSE',
    verbose=False,
    allow_writing_files=False
)


    model.fit(
        X_tr, residuals_tr,       # Train: X ve Train Hataları
        cat_features=cat_features_indices, # Varsa kategorik sütunlar
        eval_set=(X_val, residuals_val),   # DÜZELTME: Hedef olarak residuals_val verdik
        early_stopping_rounds=50
    )
    
    # Residual Tahminleri (CatBoost orijinal X_val ve test df kullanır)
    p_resid_val = model.predict(X_val)
    p_resid_test = model.predict(test)
    
    # --- F. BİRLEŞTİRME ---
    # Final Tahmin = Linear Tahmin + CatBoost'un Hata Düzeltmesi
    fold_preds_val = p_lin_val + p_resid_val
    fold_preds_test = p_lin_test + p_resid_test
    
    # --- G. SONUÇLARI SAKLAMA ---
    cat_oof_preds[val_idx] = fold_preds_val
    cat_test_preds += fold_preds_test
    
    # Skor Hesapla
    fold_rmse = np.sqrt(mean_squared_error(y_val, fold_preds_val))
    cat_scores.append(fold_rmse)
    
    print(f"Fold {fold+1} RMSE: {fold_rmse:.4f}")

# --- SONUÇLAR ---
print(f"\nOrtalama RMSE: {np.mean(cat_scores):.4f}")
print(f"Standart Sapma: {np.std(cat_scores):.4f}")

# Test tahminlerinin ortalamasını al
final_test_predictions = cat_test_preds / 10
final_test_predictions = np.clip(final_test_predictions, 0, 100)
```

```python
from sklearn.model_selection import KFold
from xgboost import XGBRegressor





kf = KFold(n_splits=10, shuffle=True, random_state=42)


#X=df.drop(columns="exam_score",axis=1)
#y=df["exam_score"]

# Sonuçları saklayacağımız boş diziler
xgb_oof_preds = np.zeros(len(X))          # Eğitim setindeki tahminler (Validation skorunu ölçmek için)
xgb_test_preds = np.zeros(len(test))   # Kaggle test seti tahminleri (Her fold'un ortalaması alınacak)
xgb_rmse_scores = []


# X ve y'nin indexlerini sıfırlayalım ki hata almayalım
X = X.reset_index(drop=True)
y = y.reset_index(drop=True)

print("K-Fold Eğitimi Başlıyor...\n")

for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
    
    # --- A. VERİYİ BÖLME ---
    X_tr, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_tr, y_val = y.iloc[train_idx], y.iloc[val_idx]
    
    # --- B. SCALING (ÖNEMLİ: Sadece Train setine fit edilir!) ---
    scaler = StandardScaler()
    X_tr_scaled = scaler.fit_transform(X_tr)
    X_val_scaled = scaler.transform(X_val)
    kagtest_scaled = scaler.transform(test) # Test verisini de bu fold'un scaler'ı ile dönüştür
    
    # --- C. MODEL 1: LINEAR REGRESSION ---
    #lin_model = LinearRegression()
    #lin_model.fit(X_tr_scaled, y_tr)
    
    # Linear Tahminler
    p_lin_train = lin_model.predict(X_tr_scaled) # Residual hesaplamak içinq
    p_lin_val = lin_model.predict(X_val_scaled)
    p_lin_test = lin_model.predict(kagtest_scaled)
    
    # --- D. RESIDUAL HESAPLAMA ---
    residuals_tr = y_tr - p_lin_train

    #'colsample_bytree': np.float64(0.7615344684232164), 'gamma': np.float64(0.03244612355449078), 'learning_rate': np.float64(0.03039154139343447), 'max_depth': 6, 'min_child_weight': 7, 'n_estimators': 2136, 'reg_alpha': np.float64(0.2848404943774676), 'reg_lambda': np.float64(1.1106608420635984), 'subsample': np.float64(0.8438257335919588

    
    # --- E. MODEL 2: XGBOOST (Residuals üzerine) ---
    # Not: XGBoost için scale edilmemiş orijinal X verisini kullanabilirsin (senin önceki kodun gibi)
    xgb_resid = XGBRegressor(
    colsample_bytree=0.7615344684232164,
    gamma=0.03244612355449078,
    learning_rate=0.03039154139343447,
    max_depth=6,
    min_child_weight=7,
    n_estimators=2136,
    reg_alpha=0.2848404943774676,
    reg_lambda=1.1106608420635984,
    subsample=0.8438257335919588,
    random_state=42,
    n_jobs=-1
)
    
    xgb_resid.fit(X_tr, residuals_tr)
    
    # Residual Tahminleri
    p_resid_val = xgb_resid.predict(X_val)
    p_resid_test = xgb_resid.predict(test)
    
    # --- F. BİRLEŞTİRME ---
    # Validation için: Linear + Residual
    fold_preds_val = p_lin_val + p_resid_val
    
    # Test (Submission) için: Linear + Residual
    fold_preds_test = p_lin_test + p_resid_test
    
    # --- G. SONUÇLARI SAKLAMA ---
    # Bu fold'un validation tahminlerini genel listeye ekle
    xgb_oof_preds[val_idx] = fold_preds_val
    # Test tahminlerini topla (döngü bitince 5'e böleceğiz)
    xgb_test_preds += fold_preds_test
    
    # Skor Hesapla
    fold_rmse = np.sqrt(mean_squared_error(y_val, fold_preds_val))
    xgb_rmse_scores.append(fold_rmse)
    
    print(f"Fold {fold+1} RMSE: {fold_rmse:.4f}")

# --- SONUÇLAR ---
print(f"\nOrtalama RMSE: {np.mean(xgb_rmse_scores):.4f}")
print(f"Standart Sapma: {np.std(xgb_rmse_scores):.4f}")

# Test tahminlerinin ortalamasını al
final_test_predictions = xgb_test_preds / 10

# 0-100 arasına sıkıştır (Sınav notu mantığı)
final_test_predictions = np.clip(final_test_predictions, 0, 100)
```

```python
# --- DÜZELTİLMİŞ BLENDING KODU ---

weight_cat = 0.45 
weight_xgb = 0.55 

# ÖNEMLİ DÜZELTME: Toplamları 10'a bölüp ortalamayı buluyoruz
# Çünkü K-Fold (10 split) kullandık ve sonuçları sürekli topladık.
avg_cat_preds = cat_test_preds / 10
avg_xgb_preds = xgb_test_preds / 10

# Final Tahmin (Ortalamalar üzerinden yapılıyor)
final_ensemble_preds = (avg_cat_preds * weight_cat) + (avg_xgb_preds * weight_xgb)

# 0-100 arasına sıkıştırma
final_ensemble_preds = np.clip(final_ensemble_preds, 0, 100)

print(f"Blend işlemi tamamlandı! (Cat: {weight_cat}, XGB: {weight_xgb})")

# --- Kaydetme ---
submission = pd.DataFrame({
    'id': test_org["id"],
    'exam_score': final_ensemble_preds
})

submission['exam_score'] = submission['exam_score'].round(1)
submission.to_csv('submission.csv', index=False)

print("Dosya hazır: submission_ensemble_fixed.csv")
print(submission.head())
```
